```python
import pandas as pd

# Load the dataset
file_path = "/kaggle/input/global-forest-data-2001-2022/deforestation/Country tree cover loss.csv"
df = pd.read_csv(file_path)

# Show first 5 rows
df.head()
```

```python
import matplotlib.pyplot as plt

# Select only the tree cover loss columns
loss_cols = [col for col in df.columns if "tc_loss" in col]

# Sum loss across all countries for each year
global_loss = df[loss_cols].sum()

# Convert to a DataFrame with years
global_loss.index = [int(col.split("_")[-1]) for col in global_loss.index]

plt.figure(figsize=(12,6))
plt.plot(global_loss.index, global_loss.values, marker='o')
plt.title("Global Tree Cover Loss (2001–2022)")
plt.xlabel("Year")
plt.ylabel("Tree Cover Loss (ha)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select only the columns with yearly losses
loss_cols = [col for col in df.columns if "tc_loss_ha_" in col]

# Sum total forest loss (2001–2022) for each country
country_loss = df.groupby("country")[loss_cols].sum().sum(axis=1)

# Get Top 5 countries
top5_loss = country_loss.sort_values(ascending=False).head(5)

# Plot
plt.figure(figsize=(10,6))
top5_loss.plot(kind='bar', color='forestgreen')
plt.title("Top 5 Countries with Highest Tree Cover Loss (2001–2022)")
plt.ylabel("Tree Cover Loss (hectares)")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 4: Calculate Net Forest Change

# First calculate total loss for each country (2001–2022)
loss_cols = [col for col in df.columns if col.startswith("tc_loss_ha_")]
df["total_loss_ha"] = df[loss_cols].sum(axis=1)

# Net change = gain - total loss
df["net_change_ha"] = df["gain_2000-2020_ha"] - df["total_loss_ha"]

# Group by country (sum across thresholds if multiple exist)
net_change = df.groupby("country")[["gain_2000-2020_ha", "total_loss_ha", "net_change_ha"]].sum()

# Top 10 countries with highest net LOSS
top_net_loss = net_change.sort_values("net_change_ha").head(10)

# Top 10 countries with net GAIN
top_net_gain = net_change.sort_values("net_change_ha", ascending=False).head(10)

top_net_loss, top_net_gain
 
```
